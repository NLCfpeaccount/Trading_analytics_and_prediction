```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Import numpy for numerical operations
import numpy as np
# Import seaborn for statistical data visualization
import seaborn as sns
# Import matplotlib.pyplot for creating static, interactive, and animated visualizations
import matplotlib.pyplot as plt
```

```python
# Read the fear and greed index data from a CSV file into a pandas DataFrame
data1= pd.read_csv("fear_greed_index.csv")
data2= pd.read_csv("historical_data.csv")
```

```python
data1.info()
```

```python
"""
1. `data1["timestamp"] = pd.to_datetime(data1["timestamp"])` - 
This line converts the "timestamp" column to datetime format using pandas' `to_datetime()` function.
This allows for easier time-based operations and analysis.

2. `data1["date"] = pd.to_datetime(data1["date"])` - 
Similarly, this line converts the "date" column to datetime format.

Converting strings to datetime objects enables powerful datetime functionality like filtering by date ranges, 
extracting components (year, month, day), resampling time series data, and calculating time differences.

"""


data1["timestamp"] = pd.to_datetime(data1["timestamp"])
data1["date"]      = pd.to_datetime(data1["date"])
```

```python
# Convert timestamp column to string format with hours, minutes, seconds and microseconds
data1["timestamp"] = data1["timestamp"].dt.strftime("%H:%M:%S:%f")

# Display the dataframe after conversion
data1
```

```python
# Display the DataFrame 'data2' to view its contents
data2
```

```python
# Convert 'Timestamp IST' column to datetime format using mixed format parsing
data2["Timestamp IST"] = pd.to_datetime(data2["Timestamp IST"],format="mixed")

# Convert 'Timestamp' column to datetime format using default parsing
data2["Timestamp"] = pd.to_datetime(data2["Timestamp"])
```

```python
"""  The selected code performs data preparation and merging operations. Here's a step-by-step explanation:

1. `data2["date"] = data2["Timestamp IST"].dt.normalize()` - 
This creates a new column called "date" in the data2 DataFrame by extracting just the date part (without time) from 
the "Timestamp IST" column.The `.dt.normalize()` method converts datetime values to midnight (00:00:00) of the same day.

2. `comdat = pd.merge(data1, data2, on="date", how="inner")` - 
This merges two DataFrames (data1 and data2) based on the "date" column. The "inner" 
join keeps only rows where the date exists in both DataFrames.

3. `comdat.drop(columns=["Timestamp", "timestamp"])` - 
This attempts to remove the columns named "Timestamp" and "timestamp" from the merged DataFrame. 
However, this line won't actually modify the DataFrame because the result isn't being assigned back to a variable 
(it should be `comdat = comdat.drop(...)` to take effect).

"""
# Convert the timestamp to date (removing time component) and create a new column 'date'
data2["date"] = data2["Timestamp IST"].dt.normalize()

# Merge data1 and data2 dataframes on the 'date' column with inner join
# (keeping only rows where dates match in both dataframes)
comdat = pd.merge(data1, data2, on="date", how="inner")

# Attempt to drop 'Timestamp' and 'timestamp' columns from the merged dataframe
# Note: This line has no effect as it doesn't assign the result back to comdat
comdat.drop(columns=["Timestamp", "timestamp"])
```

```python
# Print the shapes (dimensions) of three dataframes:
# - data1: first dataframe
# - data2: second dataframe
# - comdat: combined/processed dataframe
print(data1.shape,
data2.shape,
comdat.shape)
```

```python
# Count the number of unique accounts in the comdat dataframe
comdat["Account"].nunique()

# Calculate net profit and loss by subtracting fees from closed profit/loss
comdat["netPnL"] = comdat['Closed PnL'] - comdat['Fee']

# Display the comdat dataframe with the newly added netPnL column
comdat
```

```python
# Group the 'comdat' DataFrame by 'Account' and 'date' columns, then calculate:
dat=comdat.groupby(["Account","date"]).aggregate(
    PnL = ("netPnL","sum"),                                # Sum of net profit/loss for each account-date group
    Average_trade_size=("Size USD","mean"),                # Average trade size in USD
    trades_per_day=("Account","count"),                    # Count of trades per day
    Sentiment= ("classification",lambda x:x.mode()[0]),    # Most common sentiment classification
    win_rate=('netPnL', lambda x: (x > 0).mean())          # Percentage of profitable trades
)
dat
```

```python
# Create a bar plot showing the relationship between sentiment and win rate
sns.barplot(x=dat["Sentiment"], y=dat["win_rate"])

# Save the figure to the specified path
plt.savefig("graphs_and_tables/winrate_vs_sentiment")

# Display the plot
plt.show()
```

```python
# Create a bar plot showing the relationship between sentiment and trades per day
sns.barplot(x=dat["Sentiment"],y=dat["trades_per_day"])

# Save the generated plot to the specified file path
plt.savefig("graphs_and_tables/trades_per_day_vs_sentiment")
```

```python
# Group the data by "Sentiment" column and calculate the mean of "trades_per_day" for each sentiment category
dat1=dat.groupby("Sentiment").agg(average_trades_per_day=("trades_per_day","mean"))

# Save the grouped data to a CSV file in the "graphs_and_tables" directory
dat1.to_csv("graphs_and_tables/grouped_data3.csv")
```

```python
# Create a histogram showing the distribution of trades per day
# Using binwidth of 10 and limiting the x-axis range to 0-500
sns.histplot(x=dat["trades_per_day"], stat="count", binwidth=(10), binrange=(0,500))
# Save the histogram to a file in the graphs_and_tables directory
plt.savefig("graphs_and_tables/trades_per_day_distribution")
```
